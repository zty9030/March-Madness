```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# load data
* misc 
* summary_pt

```python
kenpom_2016 = pd.read_csv('data/2016/kenpom_2016.csv')
kenpom_2016.columns =[i.lower() for i in kenpom_2016.columns.tolist()]
del kenpom_2016['season_y']
```

```python
kenpom_2016.head()
```

# add team_id

```python
ken_lookup_id = pd.read_csv('data/ken_pom_data/kenpom_ids.csv')
```

```python
#ken_lookup_id.columns.values[2]='teamname'
ken_lookup_id.head()
```

```python
kenpom_2016['team_id'] = pd.merge(kenpom_2016,ken_lookup_id,left_on='teamname',right_on='kenpom_id',how='left')['team_id']
```

```python
kenpom_2016 = kenpom_2016.dropna()
```

# add team_id according to team name 

```python
lookup = pd.read_csv('data/master_lookup.csv')
lookup.head()
```

```python
kenpom_2016_id = pd.merge(kenpom_2016,lookup[['id_2014','id_2015']],left_on='team_id',right_on='id_2014',how='inner')
```

```python
kenpom_2016_id = kenpom_2016_id.drop(['team_id','id_2014'],axis=1)
```

```python
kenpom_2016_id.tail()
```

# merge estimated data to results dataset

```python
result = pd.read_csv('data/2016/final_part_2016.csv')
```

```python
result.head()
```

```python
team1_data = kenpom_2016_id.copy()
team1_data.columns = ['team1_'+i for i in team1_data.columns]
team2_data = kenpom_2016_id.copy()
team2_data.columns = ['team2_'+i for i in team2_data.columns]
```

```python
team1_data.head()
```

```python
result_merge_team1 = pd.merge(result,team1_data,left_on='team1_id',right_on='team1_id_2015',how='left')
```

```python
result_merge_team1.info()
```

```python
result_merged_team1_team2 = pd.merge(result_merge_team1,team2_data,left_on=['team2_id'],right_on=['team2_id_2015'],how='left')
```

```python
result_merged_team1_team2.columns
```

# ready to output

```python
col_out = []
for i in kenpom_2016.columns.tolist():
    col_out.append('team1_'+i)
    col_out.append('team2_'+i)
col_out = col_out[:-2]
```

```python
out_df = result_merged_team1_team2[col_out]
out_df.head()
```

```python
out_df.to_csv('ken_pom_2016_final.csv',index=False)
```
